```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
df_coord=pd.read_csv("places-weighted.csv")
df_coord
```

```python
from sklearn.cluster import KMeans # Importing KMeans
# Creating an instance of KMeans to find 3 clusters
kmeans_1 = KMeans(n_clusters=3)
# Using fit_predict to cluster the dataset
X = df_coord[['Longitude','Latitude','Average Time to Spend']].values
predictions = kmeans_1.fit_predict(X)
```

```python
clustered = pd.concat([df_coord.reset_index(), 
                       pd.DataFrame({'Cluster':predictions})], 
                      axis=1)
```

```python
clustered.drop('index', axis=1, inplace=True)
```

```python
conditions = [
    clustered['Cluster'] == 0, 
    clustered['Cluster'] == 1,
    clustered['Cluster'] == 2,
     ]
choices = ['Monday', 'Tuesday', 'Wednesday' ]
clustered['Vacation Day'] = np.select(conditions, choices, default='black')
```

```python
clustered.sort_values(by=['Cluster'])
```

```python
scaler = MinMaxScaler()

scaler.fit(df_coord[['Longitude']])
df_coord['Longitude'] = scaler.transform(df_coord[['Longitude']])

scaler.fit(df_coord[['Latitude']])
df_coord['Latitude'] = scaler.transform(df_coord[['Latitude']])

scaler.fit(df_coord[['Average Time to Spend']])
df_coord['Average Time to Spend'] = scaler.transform(df_coord[['Average Time to Spend']])
```
